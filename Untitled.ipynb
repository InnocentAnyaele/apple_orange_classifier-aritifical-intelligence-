```python
train = ImageDataGenerator(rescale = 1/255)
validation = ImageDataGenerator(rescale = 1/255)

train_dataset = train.flow_from_directory('C:\\Users\\HP\\Documents\\dev projects\\python workspace\\apple and orange classifier\\data\\train',
                                          target_size= (100,100),
                                          batch_size = 3,
                                          class_mode = 'binary')

validation_dataset = train.flow_from_directory('C:\\Users\\HP\\Documents\\dev projects\\python workspace\\apple and orange classifier\\data\\validation',
                                          target_size= (100,100),
                                          batch_size = 3,
                                          class_mode = 'binary')

print (train_dataset.class_indices)

print (train_dataset.classes)
```

```python
model = Sequential()
model.add(Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(100, 100, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='sigmoid'))
# compile model
opt = SGD(lr=0.001, momentum=0.9)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
 	
model.compile(loss = 'binary_crossentropy',
              optimizer=tf.keras.optimizers.RMSprop(lr=2e-5),
              metrics = ['accuracy'])

model_fit = model.fit(train_dataset, steps_per_epoch = 3, epochs =50, validation_data = validation_dataset)
```

```python
pred_path = "C:\\Users\\HP\\Documents\\dev projects\\python workspace\\apple and orange classifier\\data\\validation\\combine"

for i in os.listdir(pred_path):
    img = image.load_img(pred_path + '\\' + i, target_size = (100,100))
    plt.imshow(img)
    plt.show()
    
    X = image.img_to_array(img)
    X = np.expand_dims(X,axis = 0)
    images = np.vstack([X])
    
    val = model.predict(images)
    if val == 0:
        print("apple")
    else: 
        print("orange")
```
